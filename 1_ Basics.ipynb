# Time Series Analysis
* Focuses on analyzing data changes that occur at equally spaced time intervals. Knowing factors that produced past results, future results can be forecasted.
* Used to forecast the stock market, economic outlook, and numerous other business and scientific based projects.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
# import chart_studio.plotly as py
# import plotly.express as px
from datetime import datetime
```

```python
# Time Series Data Preprocessing
df_ts = pd.read_csv('covid19_data.csv')

country_name = 'United States'
column_name = 'new_deaths_per_million'

df_ts = df_ts[df_ts['location']==country_name]
df_ts = df_ts.set_index('date')
df_ts = df_ts.drop(df_ts.columns.difference([column_name]), axis = 'columns')
df_ts = df_ts.asfreq('d')

# # Shifting Time Data
# df_ts.shift(1) # Shifts down
# df_ts.shift(-1) # Shifts up

# Plot Moving Averages
ndpm = df_ts[column_name]
rolling_average = df_ts[column_name].rolling(window=30).mean()
moving_average = df_ts[column_name].expanding().mean()

fig = plt.figure(figsize=(12,6), dpi=100)
p1 = sns.lineplot(x = df_ts.index, y= ndpm, data=df_ts, label = 'New Deaths per Million')
p2 = sns.lineplot(x = df_ts.index, y= rolling_average, data=df_ts, label = 'Rolling Average for a 30 day window')
p3 = sns.lineplot(x = df_ts.index, y= moving_average, data=df_ts, label = 'Moving Average')
```

# Stationarity Test
* Augmented Dickey-Fuller Test
* stationary series --> constant mean and variance
* helps check if time series is stationary or not
* Highly stationary --> Low P-Value 

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(timeseries):
    print ('Dickey-Fuller Test')
    # AIC aims to minimize corresponding information
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    print (dfoutput)
    
adf_test(df_ts)
```

# Causality test
* Granger Causality test
* tells us if one time series can help us forecast the results in another 

```python
from statsmodels.tsa.stattools import grangercausalitytests

column_name = 'new_deaths_per_million'

def countries_df(country_name, column_name):
    df_ts = pd.read_csv('covid19_data.csv')
    country_df = df_ts[df_ts['location']==country_name]
    country_df = country_df.set_index('date')
    country_df = country_df.drop(country_df.columns.difference([column_name]), axis = 'columns')
    country_df = country_df.asfreq('d')
    return country_df

country_df_1 = countries_df('United States','new_deaths_per_million')
country_df_2 = countries_df('Canada', 'new_deaths_per_million')

# Make dataframe values stationary in a series format then concat them
stationary_df_1 = country_df_1[column_name].diff().dropna().rename('Deaths in US')
stationary_df_2 =  country_df_2[column_name].diff().dropna().rename('Deaths in Canada')
concat_stationary_df = pd.concat([stationary_df_1, stationary_df_2], axis=1)

grangercausalitytests(concat_stationary_df[['Deaths in US','Deaths in Canada']], maxlag=4)

# # Low p values (<0.05) for F test --> Causality over both series
# fig = plt.figure(figsize=(12,6), dpi=100)
# p1 = sns.lineplot(x = concat_stationary_df.index, y= concat_stationary_df['Deaths in US'], data=concat_stationary_df, label = 'US')
# p2 = sns.lineplot(x = concat_stationary_df.index, y= concat_stationary_df['Deaths in Canada'], data=concat_stationary_df, label = 'Canada')
```

# Seasonality Test

```python
from statsmodels.graphics.tsaplots import month_plot, quarter_plot

# Seasonal Data: Repeats after certain intervals
flight_df = sns.load_dataset('flights')
flight_df['day'] = '01'
flight_df['month'] = pd.to_datetime(flight_df.month, format='%B').dt.month
flight_df['date'] = pd.to_datetime(flight_df[['year', 'month', 'day']])
flight_df.drop(['year','month','day'], axis=1, inplace=True)
flight_df.set_index('date', inplace=True)
flight_df = flight_df.asfreq('MS')
# flight_df = flight_df.fillna(method = 'bfill')

monthly_flight_df = flight_df['passengers'].resample(rule='M').mean()
quarterly_flight_df = flight_df['passengers'].resample(rule='Q').mean()

fig = plt.figure(figsize=(12,6))
p1 = sns.lineplot(x = flight_df.index, y=flight_df['passengers'], data=flight_df, label='Passengers')
month_plot(monthly_flight_df)
quarter_plot(quarterly_flight_df);

# Check stationarity using Augmented Dickey Fuller Test
adf_test(flight_df)
# --> Not stationary
```

# ARIMA Model
* Auto Regressive Integrated Moving Average
* AR focuses on making predictions based on past values 
* I: focuses on making the time series stationary by replacing values with the differences between previous values
* MA focuses on minimizing regression errors over time

* The ACF and PACF provide parameters for ARIMA models

* Non-seasonal ARIMA models have 3 parameters
    * P = Number of time lags of the AR model
    * D = How many times we have to difference the data 
    * Q = Represents the order of the MA model

* Positive Autocorrelation at lag 1: Use AR
* Negative Autocorrelation at lag 1: Use MA
